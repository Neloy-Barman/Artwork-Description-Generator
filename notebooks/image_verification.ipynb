## **Grabbing sub-folders**

## **Function to view images distribution**

```python
def plot_images_distribution(dic):
  labels = dic.keys()
  number_of_images = dic.values()

  plt.bar(labels, number_of_images)
  plt.xlabel('Labels')
  plt.xticks(rotation=90)
  plt.ylabel('Number of images')
  plt.title('Images distribiution')
  plt.show()
```

## **Number of images within folders**

```python
def img_numbers() -> dict: return { folder_name: len(get_image_files(folder_name)) for folder_name in img_folders}
```

## **Viewing initial images distribution**

```python
img_dict = img_numbers()
print(img_dict)
plot_images_distribution(img_dict)
```

## **Function to verify images**

```python
def image_verification(folders):
  for folder in folders:
    failed = verify_images(get_image_files(folder))
    failed.map(Path.unlink)
    # for img in failed:
    #   os.remove(img)
```

## **Verifying images**

```python
image_verification(folders = img_folders)
```

## **Images disribution after cleaning**

```python
img_dict = img_numbers()
print(img_dict)
plot_images_distribution(img_dict)
```
